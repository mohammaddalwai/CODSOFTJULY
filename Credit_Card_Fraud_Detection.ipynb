```python
legit_sample_set = legit_data.sample(n=492)
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample_set, fraud_data], axis=0)
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model_selected = LogisticRegression()
```

```python
model_selected.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
X_train_prediction = model_selected.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model_selected.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
